```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Đường dẫn tới file Excel
file_path = 'C:/Users/Admin/Downloads/HSA1/HSA/Tổng hợp dữ liệu khảo sát thí sinh dự thi HSA 2023 (edited).xlsx'
df = pd.read_excel(file_path, sheet_name='Bảng hỏi 01', header=4)
print(df)
```

```python
char_columns = ['CH1', 'CH2', 'CH3', 'CH4', 'CH5', 'CH6', 'CH7', 'CH8', 'CH9']
char_to_number = {'A': 1, 'B': 2, 'C': 3, 'D': 4}
df[char_columns] = df[char_columns].replace(char_to_number)
# Hiển thị DataFrame sau khi chuyển đổi
print(df)
```

```python
correlation_matrix = df.corr()
```

```python
plt.figure(figsize=(16, 14))  
sns.heatmap(data=correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5, linecolor='white')
```

```python
# Tính độ lệch chuẩn của tất cả các cột
standard_deviations = df.std()
# Tính tần suất của tất cả các cột
value_counts = df.apply(pd.Series.value_counts)
# Hiển thị kết quả
print("Độ lệch chuẩn của các cột:")
print(standard_deviations)
print("\nTần suất của các cột:")
print(value_counts)
```
